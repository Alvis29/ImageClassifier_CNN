## Data Preproccessing

#### 1. Set the data directory

```python
data_dir = "../data"
classes = os.listdir(data_dir) # get the list of classes
classes.sort() #sort the list of classes in ascending
for item in classes:
    print(item)
```

```python
#to get the number of classes
print(f"there are {len(classes)} clases")
```

#### 2. Explore the data 

```python
# Create the figure
figure = plt.figure(figsize=(12,12))

# Setup the parameters 
columns = 4
rows = int(len(classes)/columns +1)
count = 0 

# Loop through each subfolder
for each_class in classes:
    subfolder_path = os.path.join(data_dir, each_class)
    
    # List all files in the subfolder
    files = os.listdir(subfolder_path)
    files.sort()

    image_path = os.path.join(subfolder_path, files[0])
    img = cv2.imread(image_path)
    plt.subplot(rows, columns, count+1)
    plt.imshow(cv2.cvtColor(img , cv2.COLOR_BGR2RGB))
    plt.title(each_class)
    plt.axis("off")
    count+=1
```
